### 다했당

```python
testt = test.groupby('정류소ID').count()
testt.reset_index(level=['정류소ID'], inplace=True)
testt.describe()
```

```python
#2년치 월별 데이터이므로 24가 정상임, 일부만 조사된 데이터 날리기
listt = testt[testt['사용년월'] < 24]['정류소ID'].tolist()
listt #[105000153, 105000154, 123000648]

con = test['정류소ID'].isin(listt)
final = test[~con]
final.head()
```

```python
final.describe()
```

```python
#final.astype({'사용년월' : object})
for i in range(final.shape[0]) :
    final.iloc[i,0] = str(final.iloc[i,0])

final['year'] = final['사용년월'].str[:4]
final['month'] = final['사용년월'].str[4:]
final.head()
```

```python
d_2017 = final[final['year'] == '2017']
d_2019 = final[final['year'] == '2019']

print(d_2017.shape, d_2019.shape)
```

```python
d_2019 = d_2019[['출근', '퇴근', 'year', 'month', '정류소ID']]
d_2019.head()
```

```python
ffinal = pd.merge(left=d_2017, right=d_2019, on=['정류소ID', 'month'], how='outer')
ffinal.head()
```

```python
final1 = ffinal[['정류소ID', '정류소명', '지하철역 수', '산단', '전통시장', '출근_x', '출근_y']]
final1.rename(columns={'출근_x' : 'before', '출근_y' : 'after'}, inplace=True)
final2 = ffinal[['정류소ID', '정류소명', '지하철역 수', '산단', '전통시장', '퇴근_x', '퇴근_y']]
final2.rename(columns={'퇴근_x' : 'before', '퇴근_y' : 'after'}, inplace=True)

fffinal = pd.concat([final1, final2], ignore_index=True)
fffinal.info()

fffinal.columns = ['station_id', 'station_name', 'subway', 'complex', 'market', 'before', 'after']
fffinal.head()
```

```python
fffinal.to_csv("서울최종.csv", encoding='cp949')
```
