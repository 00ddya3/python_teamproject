```python
import pandas as pd
```

```python
data = pd.read_csv("./머구 노선별 인원 수.csv", encoding='cp949')
data.head()
```

```python
data1 = data[data['구분'] == '승차']
data2 = data[data['구분'] == '하차']

data1.nunique()
```

```python
data3 = pd.merge(left=data1, right=data2, on='노선')
data3.head()
```

```python
data3['순탑승인원'] = data3['출근_x'] - data3['출근_y']
data3 = data3[['노선', '순탑승인원']]
data3.head()
```

```python
data4 = pd.read_csv("./대구광역시_시내버스 정류소별_노선별_평균배차간격_06_18_2021.csv", encoding='cp949')
data4.head()
```

```python
#data4 = data4.groupby('노선').mean()
#data4.reset_index(level='노선', inplace=True)
data4['평균배차시간'] = round(data4['평균배차시간'], 0)
data4.head()
data4.info()
```

```python
data5 = pd.merge(left = data3, right=data4, how='inner')
data5.head()
```

```python
data5['혼잡도'] = data5['순탑승인원'] / (60/data5['평균배차시간'])
data5.head()
```

```python
data5 = data5.sort_values('혼잡도', ascending=False)
data5.head(20)
```

```python
data5.to_csv("혼잡도.csv", encoding='cp949')
```
