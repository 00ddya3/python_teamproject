```python
import pandas as pd
```

```python
data = pd.read_csv("./대구_final_test.csv", encoding='cp949')
data.head()
```

```python
data.drop(['Unnamed: 0', '위도', '경도'], axis=1, inplace=True)
data.head()
```

```python
final1 = data[['정류소아이디', '정류소명_x', '지하철역 수', '산업단지', '시장', '출근']]
final1.columns = ['station_id', 'station_name', 'subway','complex', 'market', 'before']
final2 = data[['정류소아이디', '정류소명_x', '지하철역 수', '산업단지', '시장', '퇴근']]
final2.columns = ['station_id', 'station_name', 'subway','complex', 'market', 'before']

ffinal = pd.concat([final1, final2], ignore_index=True)
ffinal.info()
```

```python
ffinal['after']=0
ffinal.head()
```

```python
ffinal.to_csv("대구최종.csv", encoding='cp949')
```
