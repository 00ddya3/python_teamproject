```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import platform
import seaborn as sns
import numpy as np


system_name = platform.system()
if system_name == 'Windows':
    plt.rc('font', family='NanumGothic')
else:
    plt.rc('font', family='AppleGothic')



k00 = pd.read_csv("archive/KPopHits2000.csv")
k01 = pd.read_csv("archive/KPopHits2001.csv")
k02 = pd.read_csv("archive/KPopHits2002.csv")
k03 = pd.read_csv("archive/KPopHits2003.csv")
k04 = pd.read_csv("archive/KPopHits2004.csv")
k05 = pd.read_csv("archive/KPopHits2005.csv")
k06 = pd.read_csv("archive/KPopHits2006.csv")
k07 = pd.read_csv("archive/KPopHits2007.csv")
k08 = pd.read_csv("archive/KPopHits2008.csv")
k09 = pd.read_csv("archive/KPopHits2009.csv")
k10 = pd.read_csv("archive/KPopHits2010.csv")
k11 = pd.read_csv("archive/KPopHits2011.csv")
k12 = pd.read_csv("archive/KPopHits2012.csv")
k13 = pd.read_csv("archive/KPopHits2013.csv")
k14 = pd.read_csv("archive/KPopHits2014.csv")
k15 = pd.read_csv("archive/KPopHits2015.csv")
k16 = pd.read_csv("archive/KPopHits2016.csv")
k17 = pd.read_csv("archive/KPopHits2017.csv")
k18 = pd.read_csv("archive/KPopHits2018.csv")
k19 = pd.read_csv("archive/KPopHits2019.csv")
k20 = pd.read_csv("archive/KPopHits2020.csv")

k00['year'] = 2000
k01['year'] = 2001
k02['year'] = 2002
k03['year'] = 2003
k04['year'] = 2004
k05['year'] = 2005
k06['year'] = 2006
k07['year'] = 2007
k08['year'] = 2008
k09['year'] = 2009
k10['year'] = 2010
k11['year'] = 2011
k12['year'] = 2012
k13['year'] = 2013
k14['year'] = 2014
k15['year'] = 2015
k16['year'] = 2016
k17['year'] = 2017
k18['year'] = 2018
k19['year'] = 2019
k20['year'] = 2020
```

```python
 k = pd.concat([k00, k01, k02, k03, k04, k05, k06, k07, k08, k09, k10, k11, k12, k13, k14, k15, k16, k17, k18, k19, k20])
```

```python
print(k)
```

```python
print(k.info())
```

```python
year_df = k[k['year'] >= 2000] 
print(year_df.mean())

print("=" * 50)

year_list = list(range(2000,2021))
print(year_list)

print("=" * 50)

avg_v = []
avg_e = []
avg_l = []
avg_a = []
avg_t = []
avg_d = []


for i in year_list:
    year_df = k[k['year'] == i]
    avg_v.append(round(year_df['valence'].mean(), 2))
    avg_e.append(round(year_df['energy'].mean(), 2))
    avg_l.append(round(year_df['loudness'].mean(), 2))
    avg_a.append(round(year_df['acousticness'].mean(), 2))
    avg_t.append(round(year_df['tempo'].mean(), 2))
    avg_d.append(round(year_df['duration_ms'].mean(), 2))
    


print("연도별 평균 valence :", avg_v)
print("=" * 50)
print("연도별 평균 energy :",avg_e)
print("=" * 50)
print("연도별 평균 loudness :",avg_l)
print("=" * 50)
print("연도별 평균 acousticness :",avg_a)
print("=" * 50)
print("연도별 평균 tempo :",avg_t)
print("=" * 50)
print("연도별 평균 duration_ms :",avg_d)

```

```python
y = avg_v
x = avg_e
size = []
for item in y:
    size.append(item*500)
plt.figure(figsize=(10,8))
plt.scatter(x, y, s=size,
           c=range(21), cmap='seismic')
plt.title('Average Energy vs Average Valence')
plt.xlabel('Average Energy')
plt.ylabel('Average Valence')
plt.colorbar()
plt.show()
```
